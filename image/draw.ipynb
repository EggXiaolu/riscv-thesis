```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 
plt.rcParams.update({
    'font.family': 'serif',  # 
    'font.size': 12,         # 
    'axes.labelsize': 14,    # 
    'axes.linewidth': 1,     # 
    'grid.color': '0.8',     # 
    'grid.linestyle': '--',  # 
    'grid.linewidth': 0.5,   # 
    'legend.frameon': False  # 
})

# 
years = [1978, 1980, 1982, 1986, 1994, 1998, 2000, 2002, 2006, 2007, 2008, 2010, 2014]
instructions = [80, 140, 145, 162, 166, 223, 293, 437, 446, 500, 670, 1048, 1338]

# 
plt.figure(figsize=(10, 6), dpi=300)

# 
plt.plot(years, instructions, 
         marker='o',         # 
         markersize=8,       # 
         linewidth=2,        # 
         color='#1f77b4',    # 
         markerfacecolor='w',# 
         markeredgewidth=2)  # 

# 
plt.xlim(1976, 2016)  # 
plt.ylim(0, 1400)
plt.xlabel("Year", labelpad=10)
plt.ylabel("Number of Instructions", labelpad=10)

# 
plt.xticks(range(1978, 2014, 4))  # 19784
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(4))  # 4
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(200)) 

# 
plt.grid(True, alpha=0.5)

# 
plt.tight_layout()

#  PDF 
plt.savefig("x86_growth.pdf", format="pdf", dpi=300, bbox_inches="tight")

# 
plt.show()
```
