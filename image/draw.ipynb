```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 设置学术风格
plt.rcParams.update({
    'font.family': 'serif',  # 使用衬线字体
    'font.size': 12,         # 基础字号
    'axes.labelsize': 14,    # 坐标轴标签字号
    'axes.linewidth': 1,     # 坐标轴线宽
    'grid.color': '0.8',     # 网格线颜色
    'grid.linestyle': '--',  # 网格线样式
    'grid.linewidth': 0.5,   # 网格线宽
    'legend.frameon': False  # 图例无边框
})

# 数据准备
years = [1978, 1980, 1982, 1986, 1994, 1998, 2000, 2002, 2006, 2007, 2008, 2010, 2014]
instructions = [80, 140, 145, 162, 166, 223, 293, 437, 446, 500, 670, 1048, 1338]

# 创建图表
plt.figure(figsize=(10, 6), dpi=300)

# 绘制折线图
plt.plot(years, instructions, 
         marker='o',         # 数据点标记
         markersize=8,       # 标记大小
         linewidth=2,        # 线宽
         color='#1f77b4',    # 颜色
         markerfacecolor='w',# 标记填充色
         markeredgewidth=2)  # 标记边缘宽度

# 设置坐标轴
plt.xlim(1976, 2016)  # 留出左右边距
plt.ylim(0, 1400)
plt.xlabel("Year", labelpad=10)
plt.ylabel("Number of Instructions", labelpad=10)

# 设置刻度
plt.xticks(range(1978, 2014, 4))  # 从1978开始每4年一个刻度
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(4))  # 强制每4年显示刻度
plt.gca().yaxis.set_major_locator(ticker.MultipleLocator(200)) 

# 添加网格
plt.grid(True, alpha=0.5)

# 优化布局
plt.tight_layout()

# 保存为 PDF 矢量图
plt.savefig("x86_growth.pdf", format="pdf", dpi=300, bbox_inches="tight")

# 显示图表
plt.show()
```
